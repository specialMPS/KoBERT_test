```python
import kss
from konlpy.tag import Mecab
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Komoran
from hanspell import spell_checker
```

```python
s = '어제도 오늘같고 오늘도 내일같고 매일매일이 똑같은 상황이야 옆에 있는 친구는 여행도 가고 신나게 잘 사는데, 나는 매일 힘들고 지치네'
```

```python
s = '배고프나 참을 수 있다'
```

```python
# 오탈자/맞춤법

spelled_sent = spell_checker.check(s)
s_spell = spelled_sent.checked
s_spell
```

```python
kss.split_sentences(s_spell)
```

```python
m = Mecab()
m.pos(s_spell)
```

```python
k = Komoran()
k.pos(s_spell)
```

```python
kk = Kkma()
kk.pos(s_spell)
```
